```python
import os
from concurrent.futures import ThreadPoolExecutor

# Function to convert the contents of a file to uppercase
def convert_to_uppercase(file_name):
    uppercase_file_name = f"uppercase_{file_name}"

    # Open the original file in read mode
    file = open(file_name, "r")

    # Read the contents of the original file and convert to uppercase
    contents = file.read().upper()

    # Close the original file
    file.close()

    # Open the new file in write mode
    uppercase_file = open(uppercase_file_name, "w")

    # Write the uppercase contents to the new file
    uppercase_file.write(contents)

    # Close the new file
    uppercase_file.close()

# Convert the contents of each file to uppercase in parallel
with ThreadPoolExecutor() as executor:
    for i in range(num_files):
        file_name = f"random_strings_{i+1}MB.txt"
        executor.submit(convert_to_uppercase, file_name)
```

```python
import os
import threading
import pytube
from moviepy.editor import *

# Set the search term and number of videos to download
search_term = "Machine Learning"
num_videos = 100

# Set the number of threads for concurrent downloading and conversion
num_threads = 5

# Create a directory to store the downloaded videos
output_dir = "machine_learning_videos"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Define a function for downloading and converting videos
def download_and_convert(video_url):
    try:
        # Download the video
        yt = pytube.YouTube(video_url)
        stream = yt.streams.first()
        video_title = yt.title
        print(f"Downloading: {video_title}")
        stream.download(output_path=output_dir)

        # Convert the video to audio
        video_path = os.path.join(output_dir, f"{yt.video_id}.mp4")
        audio_path = os.path.join(output_dir, f"{yt.video_id}.mp3")
        video_clip = VideoFileClip(video_path)
        audio_clip = video_clip.audio
        audio_clip.write_audiofile(audio_path)
        audio_clip.close()
        video_clip.close()

        # Delete the original video file
        os.remove(video_path)

        print(f"Completed: {video_title}")

    except Exception as e:
        print(f"Error processing video: {e}")

# Perform the search on YouTube
search_results = pytube.Search(search_term)

# Create a list of video URLs
video_urls = [f"https://www.youtube.com/watch?v={video.video_id}" for video in search_results.results[:num_videos]]

# Create a thread for each video to download and convert
threads = []
for i in range(0, len(video_urls), num_threads):
    batch_urls = video_urls[i:i+num_threads]
    for url in batch_urls:
        thread = threading.Thread(target=download_and_convert, args=(url,))
        threads.append(thread)
        thread.start()

# Wait for all threads to complete
for thread in threads:
    thread.join()

print("Video download and conversion to audio complete!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Set the seed for reproducibility
np.random.seed(42)

# Generate the random dataset
n_samples = 500
n_features = 10

dataset = pd.DataFrame()
for col in range(n_features):
    if col < 4:
        data = np.random.uniform(low=-10, high=10, size=n_samples)
    elif col < 8:
        data = np.random.uniform(low=10, high=20, size=n_samples)
    else:
        data = np.random.uniform(low=-100, high=100, size=n_samples)
    dataset[f'Col{col+1}'] = data

dataset
```

```python
# (i) K-Means clustering
# Determine the optimal number of clusters using the elbow method
distortions = []
max_clusters = 10  # Maximum number of clusters to consider
for k in range(1, max_clusters+1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(dataset)
    distortions.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, max_clusters+1), distortions, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortion')
plt.title('K-Means Elbow Curve')
plt.show()
```

```python
# Based on the elbow curve, choose the optimal number of clusters
optimal_clusters = 3

# Perform K-Means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters = optimal_clusters, random_state=42)
kmeans_labels = kmeans.fit_predict(dataset)

# Plot the distance metric graph (pairwise distances between samples)
dist_matrix = kmeans.transform(dataset)
plt.imshow(dist_matrix, cmap='viridis', aspect='auto')
plt.colorbar(label='Distance')
plt.xlabel('Samples')
plt.ylabel('Clusters')
plt.title('K-Means Distance Metric')
plt.show()
```

```python
# (ii) Hierarchical clustering
# Calculate the linkage matrix
linkage_matrix = linkage(dataset.values, method='ward')

# Plot the distance metric graph (dendrogram)
dendrogram(linkage_matrix)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

```python
# Determine the optimal number of clusters for Hierarchical clustering
# By observing the dendrogram, we can identify the number of clusters visually
hierarchical_optimal_clusters = 3

# Perform Hierarchical clustering with the optimal number of clusters
hierarchical_labels = fcluster(linkage_matrix, hierarchical_optimal_clusters, criterion='maxclust')

# Plot the distance metric graph (pairwise distances between samples) for Hierarchical clustering
dist_matrix_hierarchical = linkage_matrix[:, 2]
plt.plot(range(1, len(dist_matrix_hierarchical)+1), dist_matrix_hierarchical, marker='o')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Distance Metric')
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp, wilcoxon, ttest_ind, ranksums

# Set the seed of reproducubility
np.random.seed(42)

# Generate the random dataset
n_samples = 500
n_columns = 5

dataset = pd.DataFrame(np.random.uniform(low=5, high=10, size=(n_samples, n_columns)),
                       columns=[f'Column {i+1}' for i in range(n_columns)])
dataset
```

```python
# (i) Perform t-Test on each column
for col in dataset.columns:
    t_statistic, p_value = ttest_1samp(dataset[col], popmean=7.5)
    print(f'T-Test for {col}: t-statistic={t_statistic:.4f}, p-value={p_value:.4f}')
```

```python
# (ii) Perform Wilcoxon Signed Rank Test on each column
for col in dataset.columns:
    statistic, p_value = wilcoxon(dataset[col] - 7.5)
    print(f'Wilcoxon Signed Rank Test for {col}: statistic={statistic:.4f}, p-value={p_value:.4f}')
```

```python
# (iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4
column3 = dataset['Column 3']
column4 = dataset['Column 4']

t_statistic, p_value_ttest = ttest_ind(column3, column4)
statistic, p_value_ranksums = ranksums(column3, column4)

print(f'Two Sample t-Test: t-statistic={t_statistic:.4f}, p-value={p_value_ttest:.4f}')
print(f'Wilcoxon Rank Sum Test: statistic={statistic:.4f}, p-value={p_value_ranksums:.4f}')
```
